```python
import pandas as pd
precios = pd.Series([15.50, 20.00, 10.25, 45.00], name='Precios')
print("--- Precios de productos ---")
print(precios)
```

```python
stock = pd.Series([100, 250, 80], index=['Camisas', 'Pantalones', 'Zapatos'])
print("\n--- Stock de tienda ---")
print(f"Existencias de Camisas: {stock['Camisas']}")
```

```python
temp_dict = {'Lunes': 22, 'Martes': 25, 'Miércoles': 21, 'Jueves': 24}
series_temp = pd.Series(temp_dict)
print("\n--- Temperaturas semanales ---")
print(series_temp)
```

```python
# Aplicar un descuento del 15% a los precios del ejemplo 1
precios_con_descuento = precios * 0.85
print("\n--- Precios con 15% de descuento ---")
print(precios_con_descuento)
```

```python
# Filtrar días con temperatura mayor a 23 grados
dias_calurosos = series_temp[series_temp > 23]
print("\n--- Días calurosos (> 23°C) ---")
print(dias_calurosos)
```
