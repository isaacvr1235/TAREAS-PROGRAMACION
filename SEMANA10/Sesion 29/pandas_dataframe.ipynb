```python
import pandas as pd

# Definimos 10 registros originales
datos = {
    'Título': ['The Witcher 3', 'Elden Ring', 'Zelda: BotW', 'Minecraft', 'Cyberpunk 2077', 
               'Hades', 'Doom Eternal', 'Red Dead 2', 'Skyrim', 'Tetris Effect'],
    'Género': ['RPG', 'Action RPG', 'Aventura', 'Sandbox', 'RPG', 
               'Roguelike', 'Shooter', 'Acción', 'RPG', 'Puzzle'],
    'Ventas (M)': [50, 20, 30, 238, 25, 1, 5, 55, 60, 1],
    'Lanzamiento': [2015, 2022, 2017, 2011, 2020, 2020, 2020, 2018, 2011, 2018]
}

df_juegos = pd.DataFrame(datos)

# Mostrar los primeros registros
print("--- DataFrame de Videojuegos ---")
display(df_juegos)
```

```python
print("\n1. Información general:")
print(df_juegos.info())

print("\n2. Estadísticas descriptivas:")
display(df_juegos.describe())
```

```python
# Filtrar juegos con ventas > 30M
exitosos = df_juegos[df_juegos['Ventas (M)'] > 30]

print("--- Juegos con más de 30 millones en ventas ---")
display(exitosos)

# Filtrar solo juegos de género RPG
rpg_games = df_juegos[df_juegos['Género'] == 'RPG']

print("\n--- Catálogo de RPGs ---")
display(rpg_games)
```
