```python
from faker import Faker
import random
import csv

# Inicializamos Faker
fake = Faker("es_MX")

# 1. Generamos una lista de 100 canciones únicas (Simulando un catálogo)
# Usamos palabras aleatorias para crear títulos que parezcan canciones
canciones_top_100 = []
for _ in range(100):
    # Generamos un título de 2 a 5 palabras
    titulo = fake.sentence(nb_words=random.randint(2, 5)).replace(".", "").title()
    canciones_top_100.append(titulo)

# Aseguramos que sean únicas
canciones_top_100 = list(set(canciones_top_100))
# Si faltan (por duplicados borrados), rellenamos
while len(canciones_top_100) < 100:
    canciones_top_100.append(fake.bs().title())

# Lista de meses
meses = [
    "Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
    "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"
]

# Función para calcular likes de forma coherente
def generar_likes(reproducciones):
    # Los likes serán entre el 0% y el 50% de las reproducciones
    tasa_likes = random.uniform(0.0, 0.5)
    return int(reproducciones * tasa_likes)

# Configuración del archivo
n = 100000  # Número de registros
nombre_archivo = "reproducciones_musicales.csv"

# Crear archivo CSV
with open(nombre_archivo, "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)

    # Escribir encabezados
    writer.writerow(["id_transaccion", "nombre_cancion", "año", "mes", "reproducciones", "likes"])

    for _ in range(n):
        cancion = random.choice(canciones_top_100)
        año = random.randint(2019, 2024) # Años recientes
        mes = random.choice(meses)

        # Generar reproducciones (entre 10 y 1 millón)
        reproducciones = random.randint(10, 1000000)

        # Generar likes basados en las reproducciones
        likes = generar_likes(reproducciones)

        writer.writerow([
            fake.uuid4(),
            cancion,
            año,
            mes,
            reproducciones,
            likes
        ])

print(f"Archivo {nombre_archivo} generado con éxito con {n} registros.")
```
