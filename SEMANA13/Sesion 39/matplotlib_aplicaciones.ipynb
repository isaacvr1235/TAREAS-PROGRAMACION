# Aplicaciones de Matplotlib: An치lisis de Datos Reales

## 1. Carga del Dataset
Usaremos el dataset del Titanic para analizar precios, edades y clases sociales.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Cargar dataset real desde URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Limpieza r치pida: Eliminar nulos en 'Age' para que los gr치ficos no fallen
df = df.dropna(subset=['Age'])

print("--- Vista previa de los datos ---")
display(df.head())
```

## 2. Histograma: Distribuci칩n de Edades
**Pregunta:** 쮺u치l era la edad m치s com칰n entre los pasajeros?
Guardaremos este gr치fico como `histograma_edades.png`.

```python
plt.figure(figsize=(10, 6))

# Crear Histograma
# bins=30 define el n칰mero de barras
plt.hist(df['Age'], bins=30, color='skyblue', edgecolor='black', alpha=0.7)

plt.title('Distribuci칩n de Edades en el Titanic')
plt.xlabel('Edad (A침os)')
plt.ylabel('Frecuencia (Cantidad de Pasajeros)')
plt.grid(axis='y', linestyle='--', alpha=0.5)

# GUARDAR IMAGEN
plt.savefig('histograma_edades.png')
print("游닞 Imagen guardada: histograma_edades.png")

plt.show()
```

## 3. Diagrama de Dispersi칩n: Edad vs Precio del Boleto
**Pregunta:** 쯃os pasajeros mayores pagaban m치s por su boleto?
Guardaremos este gr치fico como `scatter_edad_precio.png`.

```python
plt.figure(figsize=(10, 6))

# Crear Scatter Plot
# c=df['Pclass'] colorea los puntos seg칰n la clase social (1, 2, 3)
plt.scatter(df['Age'], df['Fare'], c=df['Pclass'], cmap='viridis', alpha=0.6, edgecolors='gray')

plt.title('Relaci칩n entre Edad y Precio del Boleto')
plt.xlabel('Edad')
plt.ylabel('Precio del Boleto (춲)')
plt.colorbar(label='Clase Social (1=Alta, 3=Baja)')

# GUARDAR IMAGEN
plt.savefig('scatter_edad_precio.png')
print("游닞 Imagen guardada: scatter_edad_precio.png")

plt.show()
```

## 4. Boxplot: Precios por Clase Social
**Pregunta:** 쯈u칠 tan caros eran los boletos de Primera Clase comparados con Tercera? 쮿ay precios at칤picos?
Guardaremos este gr치fico como `boxplot_precios.png`.

```python
# Preparamos los datos: extraemos el precio de cada clase en una lista separada
precios_clase_1 = df[df['Pclass'] == 1]['Fare']
precios_clase_2 = df[df['Pclass'] == 2]['Fare']
precios_clase_3 = df[df['Pclass'] == 3]['Fare']

plt.figure(figsize=(10, 6))

# Crear Boxplot
plt.boxplot([precios_clase_1, precios_clase_2, precios_clase_3],
            labels=['1ra Clase', '2da Clase', '3ra Clase'],
            patch_artist=True, # Rellenar con color
            boxprops=dict(facecolor='lightgreen'))

plt.title('Distribuci칩n de Precios por Clase Social')
plt.ylabel('Precio del Boleto (춲)')
plt.grid(axis='y', linestyle=':', alpha=0.5)

# GUARDAR IMAGEN
plt.savefig('boxplot_precios.png')
print("游닞 Imagen guardada: boxplot_precios.png")

plt.show()
```
