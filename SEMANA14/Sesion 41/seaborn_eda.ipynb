# Exploratory Data Analysis (EDA) con Seaborn

## 1. Carga y Limpieza de Datos
Utilizaremos el dataset `penguins` disponible en Seaborn.
Eliminaremos filas con datos faltantes (`NaN`) para evitar errores en los gr치ficos.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Configurar tema visual por defecto
sns.set_theme(style="whitegrid")

# Cargar dataset
df = sns.load_dataset('penguins')

# Limpieza r치pida (eliminar nulos)
df = df.dropna()

print(f"--- Dataset de Ping칲inos ({len(df)} registros) ---")
display(df.head())
```

## 2. Relaci칩n entre Variables (Scatterplot)
**Pregunta:** 쮼xiste una relaci칩n entre el largo de la aleta y el peso del ping칲ino?
Usaremos el color (`hue`) para diferenciar por especie y ver si el patr칩n cambia.

```python
plt.figure(figsize=(10, 6))

# Gr치fico de Dispersi칩n
sns.scatterplot(
    data=df,
    x='flipper_length_mm',
    y='body_mass_g',
    hue='species',    # Color por especie
    style='sex',      # Forma del punto por g칠nero (opcional)
    s=100,            # Tama침o de los puntos
    alpha=0.7         # Transparencia
)

plt.title('Relaci칩n: Largo de Aleta vs Peso Corporal')
plt.xlabel('Largo de Aleta (mm)')
plt.ylabel('Masa Corporal (g)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') # Leyenda fuera del gr치fico

plt.show()
```

### 游댌 An치lisis del Scatterplot
* **Tendencia:** Se observa una **correlaci칩n positiva fuerte**. A medida que aumenta el largo de la aleta, aumenta la masa corporal.
* **Patr칩n por Grupos:**
    * Los ping칲inos **Gentoo** (verdes) son claramente m치s grandes y pesados que los Adelie y Chinstrap.
    * Los **Adelie** y **Chinstrap** se agrupan en valores similares (menor peso y aletas m치s cortas), siendo dif칤cil distinguirlos solo por estas dos variables.

## 3. Comparaci칩n de Distribuciones (Boxplot)
**Pregunta:** 쮺칩mo var칤a la longitud del pico (*bill length*) entre las diferentes especies? 쮿ay valores at칤picos?

```python
plt.figure(figsize=(10, 6))

# Boxplot
sns.boxplot(
    data=df,
    x='species',
    y='bill_length_mm',
    palette='Set2'
)

plt.title('Distribuci칩n del Largo del Pico por Especie')
plt.ylabel('Largo del Pico (mm)')
plt.xlabel('Especie')

plt.show()
```

### 游댌 An치lisis del Boxplot
* **Comparaci칩n:**
    * La especie **Adelie** tiene el pico significativamente m치s corto (mediana alrededor de 39mm) comparado con Chinstrap y Gentoo.
    * **Chinstrap** y **Gentoo** tienen medianas similares (cerca de 46-48mm), pero Gentoo tiene un rango intercuartil (la caja) m치s amplio.
* **Outliers:**
    * Se observa un **outlier** (punto fuera de los bigotes) en la especie Gentoo, indicando un ping칲ino con el pico inusualmente corto para su grupo.

## 4. Matriz de Correlaci칩n (Heatmap)
**Objetivo:** Visualizar matem치ticamente qu칠 variables est치n m치s conectadas entre s칤.
Primero calculamos la matriz de correlaci칩n (solo con columnas num칠ricas).

```python
plt.figure(figsize=(8, 6))

# 1. Calcular correlaciones (solo num칠ricas)
matriz_corr = df.corr(numeric_only=True)

# 2. Crear Heatmap
sns.heatmap(
    matriz_corr,
    annot=True,       # Mostrar los n칰meros en las celdas
    cmap='coolwarm',  # Mapa de color (Rojo=Positivo, Azul=Negativo)
    fmt=".2f",        # 2 decimales
    vmin=-1, vmax=1   # L칤mites de la escala de color
)

plt.title('Mapa de Calor de Correlaci칩n')
plt.show()
```

### 游댌 An치lisis del Heatmap
* **Correlaci칩n M치s Alta (0.87):** Existe una relaci칩n muy fuerte entre `flipper_length_mm` (aleta) y `body_mass_g` (peso). Esto confirma lo que vimos en el primer scatterplot: aletas grandes = ping칲inos pesados.
* **Correlaci칩n Negativa (-0.24):** Curiosamente, el largo del pico (`bill_length`) y el alto del pico (`bill_depth`) tienen una leve correlaci칩n negativa si miramos todos los datos juntos.
