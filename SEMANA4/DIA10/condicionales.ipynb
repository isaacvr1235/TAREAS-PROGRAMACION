Problema 1: Clasificaci√≥n de Nota Num√©rica (Uso de elif)
üìù Descripci√≥n
Dado un puntaje num√©rico de 0 a 10, clasifica la nota en las siguientes categor√≠as:

9 a 10: Sobresaliente

7 a 8.9: Notable

6 a 6.9: Aprobado

Menor a 6: Reprobado

üíª C√≥digo

```python
def clasificar_nota(nota):
    """
    Clasifica una nota num√©rica usando if/elif/else.
    """
    print(f"--- Clasificando nota: {nota:.1f} ---")

    if not (0.0 <= nota <= 10.0):
        print("‚ö†Ô∏è Error: La nota debe estar entre 0.0 y 10.0")
        return

    # Usamos elif para encadenar las condiciones de forma eficiente
    if nota >= 9.0:
        clasificacion = "Sobresaliente"
    elif nota >= 7.0: # Implica que es < 9.0
        clasificacion = "Notable"
    elif nota >= 6.0: # Implica que es < 7.0
        clasificacion = "Aprobado"
    else: # Implica que es < 6.0
        clasificacion = "Reprobado"
        
    print(f"Resultado: {clasificacion}")

# --- Pruebas de Ejemplo ---
clasificar_nota(9.5)
clasificar_nota(7.8)
clasificar_nota(6.2)
clasificar_nota(4.9)
```

Problema 3: Clasificaci√≥n de Persona por Edad (Condici√≥n M√∫ltiple)
üìù Descripci√≥n
Clasifica a una persona seg√∫n su edad en: Ni√±o (hasta 12), Adolescente (13 a 17), Adulto (18 a 64) o Adulto Mayor (65 o m√°s).

üíª C√≥digo

```python
def clasificar_por_edad(edad):
    """
    Clasifica a una persona en una etapa de la vida seg√∫n su edad.
    """
    print(f"--- Clasificando edad: {edad} a√±os ---")
    
    if edad < 0:
        print("‚ö†Ô∏è Error: La edad no puede ser negativa.")
        return

    if edad <= 12:
        categoria = "Ni√±o"
    elif edad <= 17: # Implica que es > 12
        categoria = "Adolescente"
    elif edad <= 64: # Implica que es > 17
        categoria = "Adulto"
    else: # Implica que es >= 65
        categoria = "Adulto Mayor"
        
    print(f"Categor√≠a: {categoria}")

# --- Pruebas de Ejemplo ---
clasificar_por_edad(8)   # Ni√±o
clasificar_por_edad(15)  # Adolescente
clasificar_por_edad(35)  # Adulto
clasificar_por_edad(70)  # Adulto Mayor
```
