# Funciones básicas

Este notebook contiene 5 funciones básicas en Python con ejemplos y pruebas.

```python
def factorial(n):
    """
    Calcula el factorial de un número entero positivo.
    
    Ejemplo:
    factorial(5) -> 120
    """
    if n < 0:
        return None
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

# Ejemplos y pruebas
print(factorial(5))
assert factorial(4) == 24
```

```python
def contar_vocales(texto):
    """
    Cuenta el número de vocales en una cadena.
    
    Ejemplo:
    contar_vocales("hola") -> 2
    """
    vocales = "aeiouAEIOU"
    return sum(1 for c in texto if c in vocales)

# Ejemplos y pruebas
print(contar_vocales("Hola Mundo"))
assert contar_vocales("aeiou") == 5
```

```python
def maximo_lista(nums):
    """
    Devuelve el valor máximo de una lista de números.
    
    Ejemplo:
    maximo_lista([1, 5, 3]) -> 5
    """
    if not nums:
        return None
    maximo = nums[0]
    for n in nums:
        if n > maximo:
            maximo = n
    return maximo

# Ejemplos y pruebas
print(maximo_lista([3, 7, 2, 9]))
assert maximo_lista([1, 2, 3]) == 3
```

```python
def es_par(n):
    """
    Determina si un número es par.
    
    Ejemplo:
    es_par(4) -> True
    """
    return n % 2 == 0

# Ejemplos y pruebas
print(es_par(10))
assert es_par(3) == False
```

```python
def promedio_lista(nums):
    """
    Calcula el promedio de una lista de números.
    
    Ejemplo:
    promedio_lista([2, 4, 6]) -> 4.0
    """
    if not nums:
        return 0
    return sum(nums) / len(nums)

# Ejemplos y pruebas
print(promedio_lista([10, 20, 30]))
assert promedio_lista([2, 4, 6]) == 4.0
```
