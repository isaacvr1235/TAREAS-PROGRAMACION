```python
def division_segura():
    try:
        # Intentamos convertir la entrada a flotante
        numero = float(input("Ingresa un n√∫mero para dividir 100 entre √©l: "))
        resultado = 100 / numero
    except ValueError:
        print(" Error: Debes ingresar un n√∫mero v√°lido (no letras).")
    except ZeroDivisionError:
        print(" Error: No se puede dividir entre cero.")
    else:
        # Se ejecuta solo si no hubo errores
        print(f" ¬°√âxito! El resultado es: {resultado}")

division_segura()
```

```python
def leer_configuracion(nombre_archivo):
    try:
        with open(nombre_archivo, "r", encoding="utf-8") as archivo:
            contenido = archivo.read()
            print("Contenido del archivo:")
            print(contenido)
    except FileNotFoundError:
        print(f"‚ùå Error: El archivo '{nombre_archivo}' no se encontr√≥ en la carpeta.")
    except PermissionError:
        print("‚ùå Error: No tienes permisos para leer este archivo.")
    finally:
        print("üîç Finalizando intento de lectura de archivo.")

# Prueba con un archivo inexistente
leer_configuracion("config_secreta.txt")
```

```python
def registrar_usuario():
    print("--- Registro de Sistema ---")
    try:
        edad = int(input("Ingresa tu edad: "))
        
        if edad < 0:
            # Lanzamos una excepci√≥n manual con un mensaje personalizado
            raise ValueError("La edad no puede ser un n√∫mero negativo.")
        
        if edad < 18:
            print("Aviso: El usuario es menor de edad.")
        else:
            print("Aviso: El usuario es mayor de edad.")
            
    except ValueError as e:
        # Captura tanto el error de letras como nuestro 'raise'
        print(f" Error de validaci√≥n: {e}")
    
    else:
        # Se ejecuta si el 'try' fue exitoso
        print(" Registro procesado sin errores cr√≠ticos.")
        
    finally:
        # Se ejecuta SIEMPRE, haya error o no
        print(" Limpiando datos temporales... Proceso terminado.")

registrar_usuario()
```
