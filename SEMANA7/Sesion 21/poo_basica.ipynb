```python
class Coche:
    # El constructor inicializa los atributos de cada coche
    def __init__(self, marca, modelo, a√±o):
        self.marca = marca
        self.modelo = modelo
        self.a√±o = a√±o
        self.encendido = False # Atributo por defecto

    # M√©todo para cambiar el estado del coche
    def arrancar(self):
        if not self.encendido:
            self.encendido = True
            print(f"üöó El {self.marca} {self.modelo} ha arrancado. ¬°Rum-rum!")
        else:
            print(f"‚ö†Ô∏è El coche ya estaba encendido.")

    def detener(self):
        self.encendido = False
        print(f"üõë El {self.marca} se ha detenido.")

# --- Instanciaci√≥n y uso ---
mi_coche = Coche("Toyota", "Corolla", 2023)
print(f"Coche: {mi_coche.marca} {mi_coche.modelo} ({mi_coche.a√±o})")
mi_coche.arrancar()
```

```python
class Empleado:
    def __init__(self, nombre, salario):
        self.nombre = nombre
        self.salario = salario

    def mostrar_info(self):
        return f"Empleado: {self.nombre} | Salario actual: ${self.salario:,.2f}"

    # M√©todo para aplicar un aumento
    def aplicar_aumento(self, porcentaje):
        if porcentaje > 0:
            aumento = self.salario * (porcentaje / 100)
            self.salario += aumento
            print(f"‚úÖ Se ha aplicado un aumento del {porcentaje}% a {self.nombre}.")
        else:
            print("‚ùå El porcentaje debe ser mayor a cero.")

# --- Instanciaci√≥n y uso ---
emp1 = Empleado("Laura Garc√≠a", 25000)
print(emp1.mostrar_info())

# Aplicamos un aumento del 10%
emp1.aplicar_aumento(10)
print(emp1.mostrar_info())
```
