```python
import numpy as np

# Configuración inicial
np.random.seed(7)
matriz_10x10 = np.random.randint(1, 101, (10, 10))

print("--- 1. ANÁLISIS DE DATASET 10x10 ---")
print(f"Media Global: {np.mean(matriz_10x10):.2f}")
print(f"Varianza Global: {np.var(matriz_10x10):.2f}")
print(f"Desviación Estándar Global: {np.std(matriz_10x10):.2f}")

# Estadísticas por Ejes
print("\nEstadísticas por Columnas (axis=0):")
print(f"Mínimos: {np.min(matriz_10x10, axis=0)}")
print(f"Máximos: {np.max(matriz_10x10, axis=0)}")

print("\nEstadísticas por Filas (axis=1):")
print(f"Media: {np.mean(matriz_10x10, axis=1)}")
```

```python
print("--- 2. CÁLCULO DE PERCENTILES ---")

# Usaremos un dataset más grande para que los percentiles sean significativos
datos_p = np.random.randn(1000)
p_valores = [10, 25, 50, 75, 90]
percentiles = np.percentile(datos_p, p_valores)

for p, valor in zip(p_valores, percentiles):
    print(f"Percentil {p}: {valor:.4f}")

print("\nInterpretación:")
print(f"El 50% de los datos (mediana) es {percentiles[2]:.4f}.")
print(f"El 90% de los datos se encuentra por debajo de {percentiles[4]:.4f}.")
```

```python
print("--- 3. EXPERIMENTO DE CORRELACIÓN ---")

# Generar variables
A = np.linspace(1, 100, 100)
B = 2 * A + np.random.normal(0, 5, 100) # Dependencia lineal de A con un poco de ruido
C = np.random.rand(100) # Ruido puro sin relación

# Crear matriz de datos para correlación
stack = np.vstack([A, B, C])
matriz_corr = np.corrcoef(stack)

print("Matriz de Correlación 3x3:")
print(matriz_corr)

print("\nComentario:")
print(f"Relación A-B: {matriz_corr[0,1]:.4f} (Cercana a 1, correlación positiva muy fuerte).")
print(f"Relación A-C: {matriz_corr[0,2]:.4f} (Cercana a 0, no hay relación lineal).")
```
